```python
from ultralytics import YOLO

# Load your model
model = YOLO("yolo12n.pt")

# Export to OpenVINO format (this creates a folder)
model.export(format="openvino")
```

```python
# Load the exported OpenVINO model
ov_model = YOLO("yolo12n_openvino_model/")

# Run inference specifically on the NPU
results = ov_model("image.png", device="intel:cpu")
```
