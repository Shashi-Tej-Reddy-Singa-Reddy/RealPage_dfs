```python
import pandas as pd

df = pd.read_csv('screeningapplicant.csv')
# Clean columns
df.columns = [c.strip() for c in df.columns]
if 'Table_Name' not in df.columns:
    df.columns = ['Table_Name', 'Column_Name'] + list(df.columns[2:])

unique_tables = df['Table_Name'].nunique()
unique_columns = df['Column_Name'].nunique()
# Count occurrences of each column
col_counts = df['Column_Name'].value_counts()
shared_cols = col_counts[col_counts > 1]

print(f"Unique Tables: {unique_tables}")
print(f"Unique Columns: {unique_columns}")
print(f"Shared Columns (appear in >1 table): {len(shared_cols)}")
print("Top 5 shared columns:")
print(shared_cols.head())
```
